```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
```

```python
X = np.random.randn(1000)
a = 1.7
Xa = X*(np.abs(X)>a) - X*(np.abs(X)<=a)
data = {"X" : X, "Xa": Xa}
```

If $X$ is a gaussian vector, then all its marginal $X_i$ are 1D gaussian variables.
However, if $X$ is a multivariate random variable whose marginal are gaussians, $X$ is not necessarily gaussian.

A classical example. Let $X$ be a random variable, and let $X_a = X\times 1_{(|X| > a)} - X\times 1_{(|X| \leqslant a)}$

Both $X$ and $X_a$ are gaussians but $(X, X_a)$ is not a gaussian vector.
Indeed, $X + X_a$ won't fill the set $(0, a)$.

Vizualisation below.

```python
p = seaborn.jointplot(data=data, x="X", y="Xa")
p.fig.suptitle("Vizualisation example of a non-gaussian vector whose marginal are gaussians")
p.fig.tight_layout()
p.fig.subplots_adjust(top=0.95)
```
